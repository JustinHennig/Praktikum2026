```python
import sqlite3

connection = sqlite3.connect("test.db")
cursor = connection.cursor()
cursor.execute("CREATE TABLE IF NOT EXISTS example (id INTEGER, name TEXT, age INTEGER)")
cursor.execute("INSERT INTO example VALUES (1, 'alice', 20)")
cursor.execute("INSERT INTO example VALUES (2, 'bob', 25)")
cursor.execute("INSERT INTO example VALUES (3, 'charlie', 30)")
connection.commit()
```

```python
cursor.execute("SELECT * FROM example")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute("DELETE FROM example WHERE id = 2")
```

```python
cursor.execute("UPDATE example SET age = 31 WHERE id = 1")

age_var = 31
id_var = 1

cursor.execute("UPDATE example SET age = ? WHERE id = ?", (age_var, id_var))
```
