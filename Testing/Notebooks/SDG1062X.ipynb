# Testing the SDG1062X

```python
import pyvisa

rm = pyvisa.ResourceManager()
generator = rm.open_resource("USB0::0xF4EC::0x1103::SDG1XDDD805314::INSTR")
print(generator.query('*IDN?'))
```

Channels ein/ausschalten

```python
print(generator.query("C1:OUTP?"))

generator.write("C1:OUTP ON")
print(generator.query("C1:OUTP?"))
```

Wellenform einstellen

```python
generator.write("C1:BSWV WVTP, SINE")
generator.write("C1:BSWV FRQ, 1500")
generator.write("C1:BSWV OFST, 0")
generator.write("C1:BSWV AMP, 2")
```
