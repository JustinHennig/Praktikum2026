# Using a SDS and a SDG

```python
import pyvisa

rm = pyvisa.ResourceManager()
scope = rm.open_resource('USB0::0xF4EC::0x1008::SDS5XHEC6R0118::INSTR')
gen = rm.open_resource('USB0::0xF4EC::0x1103::SDG1XDDD805314::INSTR')

print(scope.query("*IDN?"))
print(gen.query("*IDN?"))
```
