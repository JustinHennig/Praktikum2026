# Testing pyvisa-sim



```python
import pyvisa

rm = pyvisa.ResourceManager("@sim")

resources = rm.list_resources()
print(resources)

inst = rm.open_resource(resources[5])
inst.read_termination = "\n"
inst.write_termination = "\n"
```

# Test one (changing voltage and current)

```python
print("IDN:", inst.query("*IDN?"))
print("Voltage (default):", inst.query(":VOLT:IMM:AMPL?"))
print("Current (default):", inst.query(":CURR:IMM:AMPL?"))
```

Change voltage and current

```python
inst.write(":VOLT:IMM:AMPL 3.000")
inst.write(":CURR:IMM:AMPL 2.000")
```

Print the changed information 

```python
print("Voltage:", inst.query(":VOLT:IMM:AMPL?"))
print("Current:", inst.query(":CURR:IMM:AMPL?"))
```
